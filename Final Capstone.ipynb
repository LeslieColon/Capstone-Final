## Import of Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import of CSV file

```python
df=pd.read_csv('https://data.cityofnewyork.us/api/views/43nn-pn8j/rows.csv?accessType=DOWNLOAD')
```

```python
## view of first 5 rows of data
df.head()
```

```python
##view of last 5 rows of data
df.tail()
```

```python
##total rows and columns of this dataset
df.shape
```

```python
##count of non-null values and datatypes per column
df.info()
```

```python
##count of null values per column shows we have a lot of missing data!
df.isnull().sum()
```

## Data Cleaning

```python
##list of columns to drop from dataset
drop_cols =['BUILDING','Community Board', 'Council District', 'Census Tract', 'BIN', 'BBL', 'NTA', 'Location Point1', 'PHONE']
```

```python
##pass list of columns to drop
df.drop(drop_cols, inplace = True, axis =1)
```

```python
##drop all missing values from dataset
df.dropna()
```

```python
##renamed columns for improved readability
df.rename(columns={'CAMIS': 'RESTAURANT ID', 'DBA': 'RESTAURANT NAME', 'BORO': 'CITY', 
                   'CUISINE DESCRIPTION': 'CUISINE TYPE'}, inplace=True)
```

```python
#confirming renamed columns
df.head()
```

```python
#new dataframe created without missing values
df2 = df.dropna()
```

```python
#confirming no missing values
df2.isnull().sum()
```

## Exploratory Analysis

```python
#counts by cuisine types shows top 5 cuisines are American, Chinese, Coffee/Tea, Pizza and Bakery Products & Desserts and 
# lower counts types in Californian, Iranian, Haute Cuisne, Basque and Armenian cuisines
df2['CUISINE TYPE'].value_counts()
```

```python
#counts by cities, shows Manhattan & Brooklyn with most restaurants and Bronx and Staten Island with least. This could be due
# to Manhattan and Brooklyn being more densely populated
df2.CITY.value_counts()
```

```python
#counts by neighborhoods, show 2nd & 3rd Avenue and Broadway in Manhattan and 5th Avenue in Brooklyn as having the most
# restaurants
df2.groupby(['STREET', 'CITY'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#cuisines by neighborhood show American cuisines and Coffee/Tea places on Broadway or Avenues 2-5 in Manhattan and Brooklyn
# as most common restaurants and Hot Dogs or Chicken being less common
df2.groupby(['CUISINE TYPE', 'CITY','STREET'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#sorted counts show those American fast food places are all located on Broadway in Manhattan
df2.groupby(['RESTAURANT NAME', 'STREET', 'CITY'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#grouping by Critical Flag show that out of the 5 fast food places, 718 Dunkin Donuts locations had Critical Flag citations 
#that can cause foodborne illness, 886 locations that were not critical and remaining fast food places Critical Flags 
#were not critical
df2.groupby(['RESTAURANT NAME', 'CRITICAL FLAG'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#grouping by Action show these restaurants had some form of Violation observed and cited but still had a Grade A 
#during an Inspection
df2.groupby(['ACTION', 'GRADE','RESTAURANT NAME'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#grouping Inspection grades show Top 5 restaurants with highest grade A are all Fast Food places such as Dunkin Donuts, 
#Starbucks, Subway, McDonalds and Popeye's on the Initial Inspection!
df2.groupby(['RESTAURANT NAME', 'INSPECTION TYPE', 'GRADE'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#Top 5 Violation Codes with best grade of A are 10F, 06D, 08A, 10B, and 06C
df2.groupby(['VIOLATION CODE', 'GRADE'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#Top 5 & Low 5 Violation Descriptions & Counts
df2.groupby(['VIOLATION CODE', 'VIOLATION DESCRIPTION'])['RESTAURANT ID'].count().sort_values(ascending=False)
```

```python
#2,247 Restaurants with good scores below or equal to 14 and Inspection grades of an A have Violations that are quite terrible! 
df2.loc[(df['SCORE'] <= 14 & (df['GRADE']=='A'))]
```

```python
#8 Restaurants with scores less than 27 and grades of B
df2.loc[(df['SCORE'] <27 & (df['GRADE']=='B'))]
```

```python
#5,282 Restaurants with Inspection Grade of C 
df2.loc[((df['GRADE']== 'C'))]
```

```python
#Overall scores by city show Queens and Bronx with highest scores while Brooklyn and Manhattan seem even. Staten Island's 
#scores are lowest. The lower the score, the better the grades.
sns.barplot(x= 'CITY', y='SCORE', data=df2, palette = 'plasma')
```

```python
#This lineplot shows the Critical Flag can exist with good grades and dips around grades Z-N. Z & P is grade pending and N 
#is not yet graded.
sns.lineplot(x='GRADE', y='CRITICAL FLAG', data=df2)
```
